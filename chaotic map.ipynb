```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load an image
image_path = r"C:\Users\Asus\python image encryption\images\pirate.tif"
original_image = cv2.imread(image_path)

# Flatten the original image to a 1D array for each channel
original_image_1d = original_image.reshape((-1, 3))

# Calculate the number of pixels
num_pixels = len(original_image_1d)

# Define your chaotic equation
def chaotic_equation(x, mu):
    return np.pi * np.cos(mu * x) + mu * np.sin(2 * np.pi * x)

# Generate a chaotic sequence for each channel
mu = 0.1  # You can adjust the value of mu
chaotic_sequence = chaotic_equation(np.arange(num_pixels), mu)

# Multiply each pixel value by its assigned chaotic sequence value for each channel
encrypted_image_1d = original_image_1d.astype(float) * chaotic_sequence[:num_pixels, np.newaxis]

# Reshape the encrypted 1D array back to the original image shape
encrypted_image = encrypted_image_1d.reshape(original_image.shape)

# Display the original image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()

# Display the encrypted image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(encrypted_image.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title('Encrypted Image')
plt.show()

# Save the encrypted image
cv2.imwrite(r'C:\Users\Asus\python image encryption\chaotic_image.jpg', encrypted_image)
print("Encrypted image saved as 'chaotic_image.jpg'")

# Decryption: Divide each pixel by its assigned chaotic sequence value for each channel
decrypted_image_1d = encrypted_image_1d / chaotic_sequence[:num_pixels, np.newaxis]

# Reshape the decrypted 1D array back to the original image shape
decrypted_image = decrypted_image_1d.reshape(original_image.shape)

# Display the decrypted image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(decrypted_image.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title('Decrypted Image')
plt.show()

# Save the decrypted image
cv2.imwrite(r'C:\Users\Asus\python image encryption\chaotic_decrypted_image.jpg', decrypted_image)
print("Decrypted image saved as 'chaotic_decrypted_image.jpg'")
```
