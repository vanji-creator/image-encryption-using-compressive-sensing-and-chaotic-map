```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def calculate_histogram(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Calculate histogram
    hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

    return hist

def plot_histogram(hist, title):
    plt.plot(hist, color='black')
    plt.title(title)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.show()

def calculate_entropy(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Calculate histogram
    hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

    # Normalize histogram
    hist_normalized = hist / sum(hist)

    # Calculate entropy
    entropy = -np.sum(hist_normalized * np.log2(hist_normalized + 1e-10))

    return entropy

# Load the images
image1 = cv2.imread(r'C:\Users\Asus\python image encryption\ogdog.jpg')
image2 = cv2.imread(r'C:\Users\Asus\python image encryption\chaotic_image.jpg')

# Calculate histograms
hist1 = calculate_histogram(image1)
hist2 = calculate_histogram(image2)

# Plot histograms
plot_histogram(hist1, "Histogram of Image 1")
plot_histogram(hist2, "Histogram of Image 2")

# Calculate entropy
entropy1 = calculate_entropy(image1)
entropy2 = calculate_entropy(image2)

print(f"Entropy of Image 1: {entropy1}")
print(f"Entropy of Image 2: {entropy2}")
```
