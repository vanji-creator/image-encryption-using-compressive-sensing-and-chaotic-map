```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def encrypt(image, key):
    return np.bitwise_xor(image, key)

def decrypt(encrypted_image, key):
    return np.bitwise_xor(encrypted_image, key)

# Load the image
original_image = cv2.imread(r"C:\Users\Asus\python image encryption\compressed_image.jpg", cv2.IMREAD_GRAYSCALE)

# Generate a random key (must be the same size as the image)
key = np.random.randint(256, size=original_image.shape, dtype=np.uint8)

# Encryption
encrypted_image1 = encrypt(original_image, key)
cv2.imwrite(r"C:\Users\Asus\python image encryption\xor_image.jpg", encrypted_image1)

# Display the original and encrypted images
plt.figure(figsize=(12, 6))
plt.subplot(121), plt.imshow(original_image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(encrypted_image1, cmap='gray'), plt.title('Encrypted Image')
plt.show()

# Convert color image to grayscale
deshuffled_image = cv2.imread(r"C:\Users\Asus\python image encryption\reconstructed_image.jpg", cv2.IMREAD_GRAYSCALE)

# Decryption
decrypted_image = decrypt(deshuffled_image, key)
cv2.imwrite(r"C:\Users\Asus\python image encryption\dexor_image.jpg", decrypted_image)

# Display the decrypted image
plt.figure(figsize=(6, 6))
plt.imshow(decrypted_image, cmap='gray')
plt.title('Decrypted Image')
plt.show()
```
